```python
import numpy as np
import pandas as pd
```

```python
## read file

data = pd.read_csv(r'C:\Users\RAC\Desktop\Data Science\Data\president_heights.csv')
```

```python
## display snipet of data

data.head()
```

```python
## turn height to np array

heights = np.array(data['height(cm)'])
```

```python
## print heights

print(heights)
```

```python
## simple statistics

print("Mean Height =", heights.mean())
```

```python
print('Height Standard Deviation =', heights.std())
```

```python
print("Minimun Height =", heights.min())
```

```python
print("Maximun Height =", heights.max())
```

```python
## percentiles

print("25th Percentile =", np.percentile(heights, 25))
```

```python
print("Median =", np.median(heights))
```

```python
print("75th percentile =", np.percentile(heights, 75))
```
